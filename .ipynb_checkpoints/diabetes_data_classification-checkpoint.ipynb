# Exploratory Data Analysis (EDA)

## 📊 Exploratory Data Analysis (EDA)

**Exploratory Data Analysis (EDA)** is the process of analyzing and summarizing datasets to understand their main characteristics, often using visual methods.

---

## 🔹 1. Univariate Analysis

**Definition:**  
Analysis of **a single variable** to understand its distribution, central tendency, spread, and outliers.

### 💡 Common Goals:
- Identify data types (categorical, numerical)
- Detect skewness or outliers
- Understand data range and frequency

### 📌 Techniques:
- **Numerical**:  
  - `mean`, `median`, `mode`, `std`, `min`, `max`
- **Visual**:  
  - Histogram  
  - Boxplot  
  - Countplot (for categorical variables)

---

## 🔹 2. Bivariate Analysis

**Definition:**  
Analysis of the **relationship between two variables**.

### 💡 Common Goals:
- Understand correlation or associations
- Identify patterns between input and output

### 📌 Techniques:
- **Numerical vs Numerical**:  
  - Scatter plot  
  - Correlation matrix / heatmap  
  - Pearson/Spearman correlation

- **Categorical vs Numerical**:  
  - Boxplot / Violin plot  
  - Grouped bar plots  
  - Pointplot / Stripplot

- **Categorical vs Categorical**:  
  - Crosstab / Contingency table  
  - Grouped bar chart  
  - Stacked bar chart

---

## 🔹 3. Multivariate Analysis

**Definition:**  
Analysis involving **more than two variables** to understand complex relationships and interactions.

### 💡 Common Goals:
- Detect multicollinearity
- Find interactions among features
- Understand how multiple variables affect the target

### 📌 Techniques:
- Pair plot (Seaborn `sns.pairplot`)
- Heatmap of correlation matrix
- Grouped boxplots with hue
- 3D scatter plots or contour plots
- Principal Component Analysis (PCA)

---

## 🧠 EDA Tips

- Always start with `.info()` and `.describe()` to get a quick overview.
- Use `.isnull().sum()` to check for missing values.
- Use `value_counts()` for categorical data distributions.
- Visualize early and often using `matplotlib`, `seaborn`, or `plotly`.

---

**Remember:** EDA is about asking questions, finding patterns, and forming hypotheses before building ML models.

# Encoding

## 🔤 Encoding Categorical Variables in Machine Learning

In Machine Learning, **encoding** is the process of converting **categorical variables** (text labels) into **numerical format**, so that algorithms can process them.

---

## ❓ Why is Encoding Needed?

ML algorithms **cannot handle strings or categories directly**.  
They need inputs in **numerical form**.

### 📋 Example:

| Color  | Shape   | → Must be encoded as → | Color_Encoded | Shape_Encoded |
|--------|---------|------------------------|---------------|----------------|
| Red    | Circle  |                        | 0             | 1              |
| Green  | Square  |                        | 1             | 0              |

---

## 🧠 Types of Encoding

---

### 1️⃣ Label Encoding

- Converts categories into **integer labels**.
- Each unique category gets a different integer.

### 2️⃣ One-Hot Encoding

- Converts each category into a separate binary column (0 or 1).
- Avoids the "ordering" issue in label encoding.

### 3️⃣ Ordinal Encoding

- Assigns ordered integers to ordered categories (e.g., Low < Medium < High).

### 4️⃣ Target Encoding (Mean Encoding)

- Replaces each category with the mean of the target variable for that category.

### 5️⃣ Frequency Encoding

- Replaces each category with its frequency (count) in the dataset.

### 6️⃣ Using .replace() method for encoding
You can use this method if:
- If you have domain knowledge and want to do domain-specific groupings.
- If you want intentionally assign an order or level.
- If you are not worried about encoding hundreds of categories (works well for small/medium categories).

## 🧭 When to Use What?

| Encoding Type       | Best For                          | Warning                          |
|---------------------|-----------------------------------|----------------------------------|
| Label Encoding       | Tree-based models (RF, XGB)       | Implies order                    |
| One-Hot Encoding     | Linear/logistic regression        | Creates many columns (sparse)    |
| Ordinal Encoding     | Ordered categories                | Only if order matters            |
| Target Encoding      | High-cardinality + regression     | Risk of data leakage             |
| Frequency Encoding   | Fast and simple                   | May not be informative enough    |

````python
import pandas as pd
import numpy as np

# Define the target column name
target_col = 'Diagnosis'  # Replace with your actual target column

# Select only numerical columns excluding the target column
numerical_cols = data.select_dtypes(include=['number']).columns.drop(target_col)

# Iterate over each numerical column and apply capping
for column in numerical_cols:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    
    # Compute the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Define the maximum and minimum thresholds for outliers
    Maximum = Q3 + 1.5 * IQR
    Minimum = Q1 - 1.5 * IQR
    
    # Use vectorized operations to replace outliers with the thresholds
    data[column] = np.where(data[column] > Maximum, Maximum, 
                          np.where(data[column] < Minimum, Minimum, data[column]))

# Print final shape of the dataset
print("Shape after outlier capping:", data.shape)


# Print before and after shape
print("Original shape:", data.shape)
print("New shape after outlier removal:", data.shape)
````

# Variance Inflation Factor (VIF) Notes

## 🧮 Handling High VIF in Regression

**Variance Inflation Factor (VIF)** is used to detect **multicollinearity** among independent variables in regression models.

---

## ❓ What is VIF?

VIF quantifies how much a feature's variance is inflated due to its correlation with other features.

**Formula:**
$$
\text{VIF}_i = \frac{1}{1 - R_i^2}
$$

Where \( R_i^2 \) is the R² value from regressing the \( i^{th} \) feature on all other features.

---

## 🔍 Interpreting VIF Values

| VIF Value | Interpretation                      |
|-----------|-------------------------------------|
| **< 5**   | No significant multicollinearity    |
| **5–10**  | Moderate correlation — caution zone |
| **> 10**  | High multicollinearity — serious issue ⚠️ |

---

## ✅ What To Do With High VIF Variables?

### 🅰️ **Remove One of the Correlated Features**
- Check the correlation matrix.
- Drop the less important/redundant variable.
- Helps simplify and stabilize the model.

### 🅱️ **Combine Features**
- Use dimensionality reduction (e.g., PCA).
- Engineer new features from existing ones.

### 🅲️ **Keep Them (With Caution)**
- Acceptable if the model is for **prediction** and performance is good.
- **Not recommended** for interpretable models (e.g., linear/logistic regression).

---

## 📌 Summary Table

| Situation                         | Recommendation                     |
|----------------------------------|------------------------------------|
| Low VIF (< 5)                    | ✅ Keep the variable                |
| Moderate VIF (5–10)              | ⚠️ Investigate further              |
| High VIF (> 10)                  | ❌ Remove or transform the feature  |

---

## 💡 Example

If both `Item_Weight` and `Item_MRP` have VIF > 10:
- Check their correlation.
- Use domain knowledge or importance scores.
- Keep the more relevant one and drop the other.

---

**Note:** VIF is especially important when using models that rely on the **assumptions of linearity and independent features** — like Linear and Logistic Regression.


## 💡 What Should You Do?

    If you're using Linear Regression for interpretation → remove/reduce highly correlated features (check VIF). If not i cause Unstable coefficients, Inflated standard errors and Poor interpretability

    If you're using Random Forest or XGBoost for prediction → keep them; maybe use feature importance later for pruning.

    If both are important → balance interpretability with performance.

# **Data Pre-processing Tips & Tricks for Machine Learning**


**Effective **data pre-processing** significantly enhances ML model performance. Here are some practical tips and tricks:**

---

###  ✅ **1. Handle Missing Data**
- Numerical data: Use mean, median, or mode imputation.
- Categorical data: Use mode or create a separate category ("Unknown").

---

###  ✅ **2. Remove or Correct Outliers**
- Use visualizations (boxplots, histograms) to detect outliers.
- Correct using methods like clipping, Winsorizing, or removing outliers if justified.

---

###  ✅ **3. Feature Scaling**
- Use **Normalization** (Min-Max Scaling) when data has no significant outliers.
- Use **Standardization** (Z-score Scaling) when data contains outliers or follows a Gaussian distribution.

---

###  ✅ **4. Encode Categorical Variables**
- Nominal categories: Use **One-Hot Encoding**.
- Ordinal categories: Use **Label Encoding** or custom ordinal mappings.

---

###  ✅ **5. Feature Selection & Extraction**
- Remove redundant or highly correlated features.
- Use dimensionality reduction techniques like PCA for large feature sets.

---

###  ✅ **6. Balance Your Dataset**
- Handle class imbalance using techniques like oversampling (SMOTE), undersampling, or class weighting.

---

###  ✅ **7. Split Data Properly**
- Typically use an 80-20 or 70-30 train-test split.
- Apply cross-validation (K-Fold) for robust model evaluation.

---

###  ✅ **8. Data Leakage Prevention**
- Ensure no data from testing leaks into training.
- Always perform scaling and transformations **after** splitting the data.

---

###  ✅ **9. Document Pre-processing Steps**
- Keep track of all transformations and scaling methods applied.
- Use reproducible workflows (pipelines) for consistency.

---

Implementing these practices will significantly enhance the accuracy, reliability, and robustness of your Machine Learning models.

# Imbalanced Data Notes

## 📘 What is Imbalanced Data?
Imbalanced data occurs when one class significantly outnumbers the other in a classification problem.

Common in real-world scenarios like:
- Fraud Detection (fraudulent transactions are rare)
- Medical Diagnosis (rare diseases vs. healthy cases)
- Spam Classification (most emails are not spam)

**Example Dataset Distribution:**
- Class 0 (Non-Fraud): 98%
- Class 1 (Fraud): 2%

A simple model predicting always “non-fraud” can still have 98% accuracy but fail to detect actual fraud cases.

---

## ❗ Why is Imbalanced Data a Problem?
1. **Model Bias:** Model becomes biased towards majority class.
2. **Poor Recall for Minority Class:** Fails to detect important cases.
3. **Misleading Accuracy:** High accuracy but poor on minority class.
4. **Unreliable Evaluation Metrics:** Accuracy alone is not enough.

---

## 🛠️ Solutions to Handle Imbalanced Data
Techniques are categorized into:

1. **Resampling Techniques (Data-Level)**
2. **Algorithmic Approaches (Model-Level)**
3. **Choosing the Right Metrics (Evaluation-Level)**
4. **Cost-Sensitive Learning (Optimization-Level)**

---

## 1. Resampling Techniques (Balancing the Data)

### A. Oversampling the Minority Class
- **What?** Increase number of minority samples by duplication or synthetic generation.
- **When to Use?** Small dataset where removing samples causes information loss.

#### 1. Random Oversampling (ROS)
- Duplicates minority class samples.
- ⚠️ May cause overfitting.

```python
from imblearn.over_sampling import RandomOverSampler
```

#### 2. Synthetic Minority Over-sampling Technique (SMOTE)
- Generates synthetic samples using K-Nearest Neighbors.
- Good when dataset is small.

```python
from imblearn.over_sampling import SMOTE
```

---

### B. Undersampling the Majority Class
- **What?** Reduces number of majority class instances.
- **When to Use?** When large dataset allows removing samples.

#### 1. Random Undersampling (RUS)
- Removes random majority class samples.
- ⚠️ May lead to loss of valuable data.

```python
from imblearn.under_sampling import RandomUnderSampler
```

#### 2. Cluster Centroids (CC)
- Replaces majority samples with centroids using K-means.
- Retains diversity.

```python
from imblearn.under_sampling import ClusterCentroids
```

---

### C. Hybrid Sampling (SMOTE + Undersampling)
- Combines SMOTE and undersampling for balance.
- ⚠️ Requires tuning.

```python
from imblearn.combine import SMOTETomek
```

---

## 2. Algorithmic Approaches (Using Class Weights)
Instead of modifying data, adjust model’s learning.

### A. Use Class Weights
Some ML models support weighting classes.

#### Logistic Regression Example
```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced')
model.fit(X, y)
```
- ✅ Simple, no data modification.
- ❌ May not always work well if imbalance is extreme.

# Scaling Notes

**Scaling Tools in Machine Learning**

Feature scaling is crucial in ML for ensuring that different features contribute equally to the model. Here are the most common scaling methods:

---

## 1️⃣ **Min-Max Scaling (Normalization)**
- Scales features between 0 and 1.
- Sensitive to outliers.

**Formula:**  
$$
X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}
$$

**When to use:**  
- No significant outliers.  
- Algorithms sensitive to feature range (e.g., Neural Networks).

---

## 2️⃣ **Standardization (Z-score Scaling)**
- Scales data to have mean = 0 and standard deviation = 1.
- Less sensitive to outliers.

**Formula:**  
$$
X_{scaled} = \frac{X - \mu}{\sigma}
$$

**When to use:**  
- Data has outliers.  
- Algorithms like PCA, SVM, or Linear Regression.

---

## 3️⃣ **Robust Scaler**
- Uses median and interquartile range (IQR).
- Highly resistant to outliers.

**Formula:**  
$$
X_{scaled} = \frac{X - \text{median}(X)}{\text{IQR}(X)}
$$

**When to use:**  
- Significant outliers present.

---

## 4️⃣ **Max-Abs Scaler**
- Scales each feature by its maximum absolute value.
- Range becomes [-1, 1].

**Formula:**  
$$
X_{scaled} = \frac{X}{|X_{max}|}
$$

**When to use:**  
- Sparse data (e.g., text classification or sparse matrices).

---

## 🔍 **Quick Comparison Table:**

| Scaler              | Handles Outliers | Range      | Use Cases                             |
|---------------------|------------------|------------|----------------------------------------|
| **Min-Max Scaler**  | ❌ No            | [0, 1]     | Neural networks, image data            |

# (1) K-Nearest Neighbors (KNN) Notes

**K-Nearest Neighbors (KNN) - Machine Learning**

K-Nearest Neighbors (KNN) is a **simple, non-parametric, lazy learning** algorithm used for **classification** and **regression**.

---

## 🔹 How KNN Works

1. Choose the number of neighbors, **k**.
2. Calculate the **distance** from the test point to all training points.
3. Select the **k closest** neighbors.
4. Predict the output:
   - **Classification**: Majority vote of neighbors.
   - **Regression**: Average of neighbor values.

---

## 📐 Common Distance Metrics

- **Euclidean Distance** (default):
  $$
  d(p, q) = \sqrt{ \sum_{i=1}^{n} (p_i - q_i)^2 }
  $$

- **Manhattan Distance**:
  $$
  d(p, q) = \sum_{i=1}^{n} |p_i - q_i|
  $$

---

## ⚙️ Key Parameters

- `k`: Number of neighbors (odd values preferred for binary classification).
- `metric`: Distance measure (e.g., `euclidean`, `manhattan`).
- `weights`: Can be `'uniform'` or `'distance'`.

---

## ✅ Advantages

- Simple and intuitive.
- No training phase (lazy learner).
- Works well with small datasets.

---

## ⚠️ Disadvantages

- **Slow for large datasets** (computes distance to all points).
- **Sensitive to irrelevant features** and **feature scaling**.
- **Curse of dimensionality** in high dimensions.

---

## 📌 Tips

- **Scale features** before using KNN (`StandardScaler`, `MinMaxScaler`).
- Use **odd values of k** to avoid ties.
- Choose `k` using **cross-validation**.

# Validation Notes

## 🧠 Machine Learning Validation Methods

---

### 📌 Introduction
Model validation is the process of evaluating how well a machine learning model performs on unseen data. Proper validation ensures the model generalizes well.

---

## 🧮 Regression-Specific Metrics

### 1. Mean Absolute Error (MAE)

**Formula:**  
$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$

**Where:**  
- $y_i$ = actual value  
- $\hat{y}_i$ = predicted value  
- $n$ = number of samples  

**Interpretation:**  
- Measures average magnitude of errors.  
- Easy to interpret and in original units.  
- Lower MAE is better.

---

### 2. Mean Squared Error (MSE)

**Formula:**  
$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

**Where:**  
- $y_i$ = actual value  
- $\hat{y}_i$ = predicted value  
- $n$ = number of samples  

**Interpretation:**  
- Penalizes large errors more than MAE.  
- Sensitive to outliers.  
- Lower MSE is better.

---

### 3. Root Mean Squared Error (RMSE)

**Formula:**  
$RMSE = \sqrt{MSE}$

**Interpretation:**  
- Same as MSE, but in original unit.  
- Easier to interpret than MSE.  
- Sensitive to outliers.  
- Lower RMSE is better.

---

### 4. R-squared (R²)

**Formula:**  
$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$

**Where:**  
- $SS_{res}$ = residual sum of squares  
- $SS_{tot}$ = total sum of squares  

**Interpretation:**  
- Proportion of variance explained by model.  
- $R^2 = 1$: perfect fit  
- $R^2 = 0$: no better than mean  
- $R^2 < 0$: worse than mean prediction

---

## 🧪 Classification-Specific Metrics

### 🔹 Confusion Matrix Components

|                            | Predicted Positive | Predicted Negative |
|----------------------------|--------------------|--------------------|
| **Actual Positive**        | True Positive (TP) | False Negative (FN)|
| **Actual Negative**        | False Positive (FP)| True Negative (TN) |

---

### 🔹 Accuracy

**Formula:**  
$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$

**Interpretation:**  
- Overall proportion of correct predictions.

---

### 🔹 Precision

**Formula:**  
$Precision = \frac{TP}{TP + FP}$

**Interpretation:**  
- How many predicted positives are actually positive.  
- High precision = few false positives.

---

### 🔹 Recall (Sensitivity)

**Formula:**  
$Recall = \frac{TP}{TP + FN}$

**Interpretation:**  
- How many actual positives were correctly predicted.  
- High recall = few false negatives.

---

### 🔹 F1-Score

**Formula:**  
$F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}$

**Interpretation:**  
- Harmonic mean of precision and recall.  
- Good for imbalanced datasets.

---

### 🔹 Specificity

**Formula:**  
$Specificity = \frac{TN}{TN + FP}$

**Interpretation:**  
- Ability to correctly predict negative class.

---

## 📈 ROC-AUC (Receiver Operating Characteristic - Area Under Curve)

- Evaluates model's ability to separate classes.
- Best used with probabilistic models and imbalanced data.

### ROC Definitions:

$TPR = \frac{TP}{TP + FN}$  
$FPR = \frac{FP}{FP + TN}$

---

### AUC Score Interpretation:

| AUC Score | Meaning           |
|-----------|-------------------|
| 1.0       | Perfect model      |
| 0.9–1.0   | Excellent          |
| 0.8–0.9   | Good               |
| 0.7–0.8   | Fair               |
| 0.6–0.7   | Poor               |
| 0.5       | Random guess       |
| < 0.5     | Worse than random  |

---

### Important:

Use probability predictions:  
`model.predict_proba(X)[:, 1]` ✅  
Avoid:  
`model.predict(X)` ❌

---

## 📉 Log Loss (Logarithmic/Cross-Entropy Loss)

**Formula:**  
$LogLoss = -\frac{1}{n} \sum_{i=1}^{n} \left[y_i \log(p_i) + (1 - y_i) \log(1 - p_i)\right]$

**Where:**  
- $y_i$ = actual class label (0 or 1)  
- $p_i$ = predicted probability of class 1  
- $n$ = number of samples  

**Interpretation:**  
- Measures error between predicted probabilities and actual classes.  
- Lower is better.  
- Penalizes confident wrong predictions.

---

## 🔬 Regression vs Classification

| Feature             | Regression               | Classification                |
|---------------------|--------------------------|-------------------------------|
| Target Variable     | Continuous               | Categorical                   |
| Metrics             | MAE, MSE, RMSE, R²       | Accuracy, Precision, Recall, F1 |
| Stratification      | Not used                 | Commonly used                 |
| ROC-AUC             | Not applicable           | Commonly used                 |

---

## 🧠 Choosing the Right Metric

| Scenario                                 | Best Metric        |
|------------------------------------------|--------------------|
| Regression with outliers                 | RMSE               |
| Small errors important                   | MAE                |
| Compare regression models                | R²                 |
| Balanced classification                  | Accuracy           |
| Imbalanced classification                | F1-Score           |
| Costly false positives                   | Precision          |
| Costly false negatives                   | Recall             |
| Probabilistic classification             | Log Loss, ROC-AUC  |

---

## ✅ Best Practices

- Use cross-validation for reliability.  
- Apply stratified sampling in classification.  
- Scale features for SVM/KNN.  
- Avoid data leakage between train and test sets.

# Cross-Validation Notes

Checking: “Will my model work well on new data?”

purpose is to tests the model on different subsets of the data to ensure it's not just performing well on one specific train-test split

imagine your KNN gets:

    R² = 0.99 on training set

    But R² = 0.40 in cross-validation

That’s a red flag 🚩 — it means your model fits training data well but struggles on unseen parts → overfitting.

### 1. Holdout Method
- Split dataset into:
  - Training set (e.g., 70%)
  - Testing set (e.g., 30%)
- Fast but less reliable on small datasets.

### 2. K-Fold Cross-Validation
- Split data into *k* equal-sized folds.
- Train on *k-1* folds, validate on the remaining.
- Repeat *k* times and average the results.

### 3. Stratified K-Fold (for Classification)
- Maintains class proportion in each fold.
- Prevents bias in imbalanced classification datasets.

### 4. Leave-One-Out Cross-Validation (LOOCV)
- Special case of K-Fold where *k = n* (number of data points).
- Good for small datasets but computationally expensive.

### 5. Repeated K-Fold CV
- Repeats K-Fold multiple times with different splits.
- Reduces variance in performance estimates.

---

# (2) Linear Regression Notes

**Linear Regression - Machine Learning**

**Linear Regression** is a **supervised learning** algorithm used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation.

---

## 📌 Goal

To predict a continuous numeric output \( y \) based on one or more input features \( x \).

---

## 📐 Equation of a Simple Linear Regression

$$
y = \beta_0 + \beta_1 x + \varepsilon
$$

Where:
- \( y \): Predicted output
- \( x \): Input (independent variable)
- \( \beta_0 \): Intercept
- \( \beta_1 \): Slope (coefficient)
- \( \varepsilon \): Error term (residual)

---

## 🔢 Multiple Linear Regression

For multiple features:

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \varepsilon
$$

---

## ⚙️ Assumptions of Linear Regression

1. **Linearity**: Relationship between features and target is linear.
2. **Independence**: Observations are independent.
3. **Homoscedasticity**: Constant variance of residuals.
4. **Normality of residuals**: Errors are normally distributed.
5. **No multicollinearity**: Features are not highly correlated with each other.

---

## ✅ Advantages

- Simple and easy to interpret.
- Computationally efficient.
- Works well with linearly separable data.

---

## ⚠️ Disadvantages

- Assumes linear relationships.
- Sensitive to outliers.
- Poor performance with non-linear data.

---

## 📊 Evaluation Metrics

- **Mean Absolute Error (MAE)**
- **Mean Squared Error (MSE)**
- **Root Mean Squared Error (RMSE)**
- **R² Score (Coefficient of Determination)**

# (3) Logistic Regression Notes

**Logistic Regression - Machine Learning**

**Logistic Regression** is a **supervised learning algorithm** used for **binary classification** (and with extensions, multi-class). Despite the name, it is a **classification** algorithm, not regression.

---

## 📌 Objective

To predict the **probability** that an instance belongs to a particular class.

---

## 📐 Logistic Function (Sigmoid)

Instead of predicting continuous values like linear regression, logistic regression maps predictions to probabilities using the **sigmoid function**:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$
 
Where:  
- $z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n$  
- Output $\sigma(z)$ is in range $(0, 1)$


---

## 🔗 Prediction Rule

- If $ \sigma(z) \geq 0.5 $, classify as **Class 1**  
- If $ \sigma(z) < 0.5 $, classify as **Class 0**


*(The threshold can be adjusted for specific use cases)*

---

## 🧮 Loss Function (Log Loss / Binary Cross Entropy)

The loss used is the **Log Loss**:

$$
\text{Loss} = -\left[y \log(\hat{y}) + (1 - y) \log(1 - \hat{y})\right]
$$

---

## ✅ Advantages

- Easy to implement and interpret.
- Outputs calibrated probabilities.
- Works well with linearly separable data.

---

## ⚠️ Disadvantages

# (4) Decision Tree Notes
Can be used for both Classification and Regression data

## **Decision Tree - Detailed Explanation**

### 🌳 **1. Introduction to Decision Trees**

A **Decision Tree** is a flowchart-like tree structure used for **classification** and **regression**. It splits a dataset into smaller and smaller subsets based on feature values, leading to decision outcomes (labels or values).

📌 **Why Decision Trees?**

- Simple to understand and interpret.
- Requires little data preprocessing (no need for feature scaling or normalization).
- Can handle both numerical and categorical data.
- Used as the base model in ensemble methods like **Random Forest** and **Gradient Boosting**.

---

### 🧠 **2. Key Components of a Decision Tree**

- **Root Node**: Represents the entire dataset and the first splitting condition.
- **Decision Node**: Splits the data based on a certain feature.
- **Leaf Node (Terminal Node)**: Final output (class label or value).
- **Branches**: Show outcomes of decisions or tests.
- **Splitting**: The process of dividing data into subsets.
- **Pruning**: Reducing the size of the tree to prevent overfitting.

---

### ⚙️ **3. How a Decision Tree Works (Step-by-Step)**

#### Step 1: **Select the Best Feature to Split**
- The feature that offers the **most information gain** or **least impurity** is selected.
- Criteria:
  - **Information Gain** (for classification)
  - **Gini Impurity** (for classification)
  - **Variance Reduction** (for regression)

#### Step 2: **Split the Dataset**
- Based on the selected feature and threshold (for numerical features) or category (for categorical features).

#### Step 3: **Repeat**
- Apply the same process recursively on each child node until:
  - All instances in a node belong to the same class.
  - Maximum depth is reached.
  - Further splitting doesn't improve performance.

---

### 🧮 **4. Mathematical Formulations**

#### 🔹 Entropy (for classification):

Measures impurity or randomness.

\[
\text{Entropy} = -\sum_{i=1}^{n} p_i \log_2(p_i)
\]

Where \( p_i \) is the probability of class \( i \).

#### 🔹 Information Gain:

\[
IG = \text{Entropy(parent)} - \sum_{k} \frac{|child_k|}{|parent|} \times \text{Entropy}(child_k)
\]

Used to select the best feature.

#### 🔹 Gini Impurity:

\[
\text{Gini} = 1 - \sum_{i=1}^{n} p_i^2
\]

Lower Gini = better split.

#### 🔹 Variance Reduction (for regression):

\[
\text{Variance Reduction} = \text{Var(parent)} - \sum_{k} \frac{|child_k|}{|parent|} \times \text{Var}(child_k)
\]

---

### ⚙️ **5. Hyperparameters in Decision Trees**

| Hyperparameter       | Description |
|----------------------|-------------|
| `max_depth`          | Max depth of the tree. Limits overfitting. |
| `min_samples_split`  | Minimum number of samples to split a node. |
| `min_samples_leaf`   | Minimum samples required at a leaf node. |
| `max_features`       | Number of features considered for best split. |
| `criterion`          | Function to measure quality of split (`gini`, `entropy`, or `mse`). |
| `splitter`           | Strategy for choosing the split (`best` or `random`). |

---

### ✅ **6. Advantages of Decision Trees**

- Simple to **visualize** and **interpret**.
- Works well with both **numerical and categorical** data.
- Requires **little preprocessing** (no need for normalization or scaling).
- Can **handle missing values** to some extent.

---

### ❌ **7. Disadvantages of Decision Trees**

- **Overfitting**: Trees can become overly complex and fit noise.
- **Instability**: Small changes in data can result in different trees.
- **Greedy Nature**: Locally optimal decisions may not lead to global optimum.
- **Biased toward features with more levels**.

---

### 📊 **8. Decision Tree in Practice**

- **Classification** Example: Predicting if a customer will churn or not.
- **Regression** Example: Predicting house prices based on features like size, location, etc.
- **Visualization**: Trees can be visualized using `plot_tree()` or `export_graphviz()` in `scikit-learn`.

# (5) Random Forest Notes
Can be used for both Classification and Regression data

Random Forest is an ensemble learning method based on decision trees. It combines multiple decision trees to produce better predictive accuracy and generalization, reducing issues like overfitting.

📌 Why Random Forest?

Individual Decision Trees can overfit (too specific to training data), leading to lower accuracy on new data. A Random Forest improves accuracy by:

    Combining multiple diverse decision trees.
    Reducing variance and overfitting through averaging results.

🌳 How Random Forest Works (Step-by-Step)

    Bootstrap Aggregation (Bagging):
        Random subsets of the dataset (sampling with replacement) are used to train each decision tree.
        Each tree sees a slightly different dataset.

    Feature Randomness:
        At every split, each tree considers only a random subset of features.
        This increases diversity among trees.

    Build Multiple Decision Trees:
        Repeat this process to create many decision trees (typically tens or hundreds).

    Aggregate Predictions:
        Classification: Majority voting.
        Regression: Average prediction.

📐 Random Forest Equations and Formulation

Random Forest predictions are aggregated from multiple decision trees:
Classification (Majority Voting):

$ \hat{y} = \text{mode}{T_1(x), T_2(x), \dots, T_n(x)} $

$\hat{y}$ = Final prediction.
$T_n(x)$ = Prediction from the nthnth tree.

Regression (Average Prediction):

$ \hat{y} = \frac{1}{n}\sum_{i=1}^{n} T_i(x) $

$\hat{y}$ = Final predicted value.
$T_i(x)$ = Prediction from the ithith tree.
$n$ = Number of trees.

🚩 Hyperparameters of Random Forest

    Number of Trees (n_estimators):
    More trees improve accuracy up to a point (commonly 50–200).

    Max Depth (max_depth):
    Limits tree growth to control overfitting.

    Minimum Samples per Leaf (min_samples_leaf):
    Ensures that each leaf has a minimum number of samples, controlling complexity.

    Max Features (max_features):
    Number of features considered at each split, increasing randomness.

📊 Feature Importance

Random Forests also measure the importance of features by averaging how much each feature decreases the impurity (e.g., Gini impurity) across all trees.

Feature Importance Equation:

    Averaging feature importance over all trees: 
$ \text{Importance}(X_i) = \frac{1}{N} \sum_{t=1}^{N} (\text{Importance of feature } X_i \text{ in tree } t) $


## **Random Forest - Detailed Explanation**  

### **1. Introduction to Random Forest**  
Random Forest is a powerful **ensemble learning** method that combines multiple decision trees to improve accuracy, reduce overfitting, and increase robustness. It can be used for both **classification** and **regression** tasks.


 Key Terminologies:

    Root Node: Represents the entire dataset and the first decision or split.
    Decision Nodes: Intermediate nodes that split data based on specific features.
    Leaf Node (Terminal Node): Nodes with final outcomes (predictions).
    Branches: Lines connecting nodes, representing possible decisions/outcomes.
    Splitting: Dividing nodes based on feature values.
    Pruning: Technique to reduce overfitting by trimming tree branches.

Measure the Quality of Splits: 
1. Entropy (Classification):

Entropy measures randomness in a dataset. Lower entropy = higher purity.

$
\text{Entropy} = -\sum_{i=1}^{n} p_i \log_{2}(p_i)
$

Where:

$p_i$ = Probability of class $i$.\
$n$ = Number of classes.
    

2. Information Gain (Classification):

Information Gain measures reduction in entropy after splitting.

$ \text{Information Gain (IG)} = \text{Entropy(parent)} - \displaystyle\sum_{k} \frac{|child_k|}{|parent|}\text{Entropy(child}_k) $

Where:

$\text{Entropy(parent)}$ = Entropy of the parent node before splitting.\
$\text{Entropy(child}_k)$ = Entropy of the $k^{th}$ child node after splitting.\
$|child_k|$ = Number of instances in the $k^{th}$ child node.\
$|parent|$ = Number of instances in the parent node.

Higher IG → better split.

3. Gini Impurity (Classification):

Measure of node impurity. Lower Gini impurity → better split.

$ \text{Gini Impurity} = 1 - \displaystyle\sum_{i=1}^{n} p_i^{2} $

Where:

$p_i$ = Probability of class $i$.
$n$ = Number of classes.


4. Variance Reduction (Regression):

$ \text{Variance Reduction} = \text{Variance(parent)} - \displaystyle\sum_{k} \frac{|child_k|}{|parent|}\text{Variance(child}_k) $

Where:

$\text{Variance(parent)}$ = Variance of the target variable in the parent node.\
$\text{Variance(child}_k)$ = Variance of the target variable in the $k^{th}$ child node.\
$|child_k|$ = Number of instances in the $k^{th}$ child node.\
$|parent|$ = Number of instances in the parent node.

Variance Reduction = reduction in variance after splitting. Higher reduction is better.


---

### **2. Why Use Random Forest?**  
- **Reduces Overfitting**: Unlike a single decision tree, Random Forest builds multiple trees and takes an average (regression) or majority vote (classification), reducing variance.  
- **Handles Missing Values & Outliers Well**: Since multiple trees are used, missing values or noise in data have minimal impact.  
- **Works Well with Large Datasets**: Can handle high-dimensional data effectively.  
- **Handles Feature Importance**: Can provide insights into which features are most important for predictions.  

---

### **3. How Random Forest Works?**  

#### **Step 1: Bootstrap Sampling (Bagging)**
- The dataset is randomly **sampled with replacement** to create multiple subsets (called bootstrap samples).  
- Each subset is used to train a separate decision tree.  
- Some data points might appear multiple times in a subset, while others might not appear at all.  

#### **Step 2: Random Feature Selection**
- Instead of considering all features while splitting nodes (like in Decision Trees), Random Forest randomly selects a subset of features at each split.
- This ensures **diversity** among trees and prevents overfitting.  

#### **Step 3: Building Multiple Decision Trees**
- Each decision tree is trained independently on a different bootstrap sample.  
- The trees are **deep but not necessarily pruned**, as overfitting is controlled by aggregation.  

#### **Step 4: Aggregation of Results**
- **For Classification**: Majority voting is used—each tree predicts a class, and the most voted class is the final output.  
- **For Regression**: The average of all tree predictions is taken as the final output.  

---

### **4. Hyperparameters in Random Forest**  
Fine-tuning hyperparameters helps improve the performance of the Random Forest model.  

| Hyperparameter | Description |
|----------------------|-------------|
| `n_estimators` | Number of trees in the forest. More trees increase accuracy but also computational cost. |
| `max_features` | Number of features considered for each split. Lower values increase diversity but may reduce accuracy. |
| `max_depth` | Maximum depth of each tree. Prevents overfitting if set properly. |
| `min_samples_split` | Minimum number of samples required to split a node. Higher values make trees simpler. |
| `min_samples_leaf` | Minimum samples required in a leaf node. Prevents small leaf nodes. |
| `bootstrap` | Whether bootstrap samples are used. Default is True. |

---

### **5. Advantages & Disadvantages**  

#### ✅ **Advantages**  
✔ **High Accuracy**: Generally performs better than individual decision trees.  
✔ **Handles Non-linearity**: Works well with complex datasets.  
✔ **Feature Importance**: Can rank feature importance automatically.  
✔ **Robust to Noise**: Less sensitive to outliers compared to a single tree.  
✔ **Scalability**: Can handle large datasets efficiently.  

#### ❌ **Disadvantages**  
✘ **Computationally Expensive**: Training multiple trees requires more time and resources.  
✘ **Black Box Model**: Hard to interpret compared to a single decision tree.  
✘ **More Memory Usage**: Storing multiple trees takes up more memory.

### 🔧 Hyperparameter Tuning in Machine Learning

Hyperparameter tuning is the process of finding the best set of parameters for a machine learning model to improve performance.

---

### 1. **GridSearchCV**

- **Library**: `sklearn.model_selection`
- **Description**: Exhaustively searches over a manually specified parameter grid.
- **Benefits**:
  - Simple and systematic.
  - Guarantees best model *within* the grid.
- **Drawbacks**:
  - Computationally expensive.
  - Inefficient with large parameter spaces.
- **Use Case**: When you have a small set of hyperparameters and enough compute power.

```python
from sklearn.model_selection import GridSearchCV
```

---

### 2. **RandomizedSearchCV**

- **Library**: `sklearn.model_selection`
- **Description**: Randomly samples a fixed number of parameter combinations.
- **Benefits**:
  - Faster than GridSearchCV.
  - Good for large search spaces.
- **Drawbacks**:
  - Might miss optimal settings.
- **Use Case**: When model training is expensive, and grid search is infeasible.

```python
from sklearn.model_selection import RandomizedSearchCV
```

---

### 3. **Bayesian Optimization (BayesSearchCV)**

- **Library**: `scikit-optimize` (`skopt`)
- **Description**: Models the function using past evaluations and selects promising combinations.
- **Benefits**:
  - Smarter than random/grid search.
  - Fewer evaluations needed.
- **Drawbacks**:
  - More complex to implement.
- **Use Case**: Optimizing models with costly training times or many hyperparameters.

```python
from skopt import BayesSearchCV
```

---

### 4. **Optuna**

- **Library**: `optuna`
- **Description**: Framework for automatic hyperparameter optimization with pruning and visualization.
- **Benefits**:
  - Lightweight and fast.
  - Good visualization tools.
- **Drawbacks**:
  - Slightly different syntax from sklearn.
- **Use Case**: Fast and scalable optimization in research or production.

```python
import optuna
```

---

### 5. **Hyperopt**

- **Library**: `hyperopt`
- **Description**: Uses Tree-structured Parzen Estimator (TPE) for probabilistic search.
- **Benefits**:
  - Flexible and scalable.
- **Drawbacks**:
  - Setup is more involved.
- **Use Case**: Non-sklearn models, deep learning tasks, large search spaces.

```python
from hyperopt import fmin, tpe, hp, Trials
```

---

### 6. **TPOT (Genetic Algorithm / Evolutionary Search)**

- **Library**: `tpot`
- **Description**: AutoML using genetic programming to evolve model pipelines.
- **Benefits**:
  - Fully automated.
  - Optimizes both model and hyperparameters.
- **Drawbacks**:
  - Long runtime.
- **Use Case**: AutoML experimentation, prototyping.

```python
from tpot import TPOTClassifier
```

---

### 7. **Successive Halving (HalvingGridSearchCV / HalvingRandomSearchCV)**

- **Library**: `sklearn.experimental`
- **Description**: Allocates resources dynamically, eliminating poor models early.
- **Benefits**:
  - Efficient with limited resources.
  - Supports early stopping.
- **Drawbacks**:
  - Experimental API in sklearn.
- **Use Case**: Situations with time/memory constraints.

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV
```

### 8. **Manual Tuning**

- **Drawbacks**:
  - Not systematic.

### 9. **AutoML Tools**

### 🔧 Common Hyperparameter Grids for GridSearchCV

Below are commonly tuned hyperparameters for popular scikit-learn models, along with short explanations.

---

### ### 🧱 K-Nearest Neighbors (KNN)

```python
param_grid = {
    'knn__n_neighbors': [3, 5, 7],
    'knn__weights': ['uniform', 'distance'],
    'knn__p': [1, 2]  # 1: Manhattan, 2: Euclidean
}
```

- `n_neighbors`: Number of neighbors to consider.
- `weights`: How neighbors are weighted.
- `p`: Distance metric (`p=1` = Manhattan, `p=2` = Euclidean).

---

### 📈 Linear Regression

```python
param_grid = {}
```

- No hyperparameters to tune in basic `LinearRegression()`.

---

### ✅ Logistic Regression (Classification)

```python
param_grid = {
    'logistic__C': [0.01, 0.1, 1, 10],
    'logistic__solver': ['liblinear', 'lbfgs']
}
```

- `C`: Inverse of regularization strength (lower = more regularization).
- `solver`: Optimization algorithm.

---

### 🌲 Decision Tree

```python
param_grid = {
    'dt__max_depth': [None, 10, 20],
    'dt__min_samples_split': [2, 5],
    'dt__min_samples_leaf': [1, 2]
}
```

- `max_depth`: Limits tree depth (None = unlimited).
- `min_samples_split`: Minimum samples to split a node.
- `min_samples_leaf`: Minimum samples in a leaf.

---

### 🌳 Random Forest

```python
param_grid = {

}
```

- `n_estimators`: Number of trees in the forest.

````python
# hyperparameter tuning using GridSearchCV — the process of finding the best model settings (like the best k in KNN) using cross-validation.
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
# from sklearn.neighbors import KNeighborsRegressor
# from sklearn.model_selection import KFold, cross_val_score # KFold - regression data
from sklearn.model_selection import StratifiedKFold, cross_val_score # StratifiedKFold - classification data
# from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
# from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
# from sklearn.tree import DecisionTreeClassifier, plot_tree
# from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, log_loss, classification_report
# from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
````

#### Specify which hyperparameters you want to tune for each model.

````python
# KNN Hyperparameter Grid
knn_param_grid = {
    'knn__n_neighbors': [3, 5, 7, 9, 11],  # double underscore required e.g: knn__
    'knn__weights': ['uniform', 'distance'],
    'knn__p': [1, 2] # 1 = Manhattan, 2 = Euclidean
}

# # Linear Regression (No hyperparameters to tune)
# linear_param_grid = {}

# Logistic Regression Hyperparameter Grid (if classification)
logistic_param_grid = {
    'logistic__C': [0.01, 0.1, 1, 10, 100], # Regularization strength
    'logistic__solver': ['lbfgs', 'liblinear']
}

# Decision Tree Hyperparameter Grid
decision_tree_param_grid = {
    'dt__max_depth': [None, 10, 20, 30, 50],
    'dt__min_samples_split': [2, 5, 10],
    'dt__min_samples_leaf': [1, 2, 5]
}

# Random Forest Hyperparameter Grid
random_forest_param_grid = {
    'rf__n_estimators': [50, 100, 200],
    'rf__max_depth': [None, 10, 20, 30],
    'rf__min_samples_split': [2, 5, 10],
    'rf__min_samples_leaf': [1, 2, 5]
}
````

#### Pipelines for models
1. Combining Preprocessing and Modeling

    Chains multiple steps (e.g., scaling, encoding, modeling) into a single object.

2. Prevents Data Leakage

    Preprocessing (like StandardScaler) is fit only on training data during cross-validation.

    Avoids mistakenly fitting scalers on test data.

````python
%%time

# Pipelines for models
knn_pipeline = Pipeline([('scaler', StandardScaler()), ('knn', KNeighborsClassifier())])
# linear_pipeline = Pipeline([('scaler', StandardScaler()), ('linear', LinearRegression())])
logistic_pipeline = Pipeline([('scaler', StandardScaler()), ('logistic', LogisticRegression(max_iter=1000))])
decision_tree_pipeline = Pipeline([('dt', DecisionTreeClassifier())])
random_forest_pipeline = Pipeline([('rf', RandomForestClassifier(random_state=42))])

# Dictionary of machine learning models with their corresponding pipelines and hyperparameter grids
# Loop over multiple models for training, comparison, or tuning.
models = {
    "KNN": (knn_pipeline, knn_param_grid),
    # "Linear Regression": (linear_pipeline, linear_param_grid),
    "Logistic Regression": (logistic_pipeline, logistic_param_grid),
    "Decision Tree": (decision_tree_pipeline, decision_tree_param_grid),
    "Random Forest": (random_forest_pipeline, random_forest_param_grid)
}


# Dictionary to store the best models after hyperparameter tuning
best_models = {}

# Loop through each model and its associated pipeline and param grid
for name, (pipeline, param_grid) in models.items():
    print(f"Tuning {name}...")
    print("="*50)
    # Initialize GridSearchCV with pipeline, param grid, 5-fold CV, and R² scoring
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='r2', n_jobs=-1)
    
    # Fit GridSearchCV on training data
    grid_search.fit(x_train, y_train)

    # Store the best estimator (i.e., the pipeline with best hyperparameters) in the dictionary
    best_models[name] = grid_search.best_estimator_

    # Print the best hyperparameters found for this model
    print(f"Best Hyperparameters for {name}: {grid_search.best_params_}")
    # Print the best cross-validation R² score
    print(f"Best Cross-validation R² Score for {name}: {grid_search.best_score_}\n")

# Train all best models on full training data
for name, model in best_models.items():
    train_score = model.score(x_train, y_train)
    # model.fit(x_train, y_train)
    print(f"Final Train Score for {name}: {train_score}")

# Evaluate each model on the test set
for name, model in best_models.items():
    test_score = model.score(x_test, y_test)
    print(f"Final Test R² Score for {name}: {test_score}")

# Dictionary to store validation metrics
validation_results = {}
actual_vs_predicted = {}

for name, model in best_models.items():
    y_pred = model.predict(x_test)

    # Compute classification metrics
    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    precision = report['weighted avg']['precision']
    recall = report['weighted avg']['recall']
    f1 = report['weighted avg']['f1-score']

    validation_results[name] = {
        "Accuracy": acc,
        "Precision": precision,
        "Recall": recall,
        "F1-score": f1
    }

    # Store actual vs predicted for visualization
    actual_vs_predicted[name] = pd.DataFrame({
        "Actual": y_test.values.copy(),
        "Predicted": y_pred
    })

    # Print actual vs predicted
    print('------------------------------------------------------')
    print(f"\n{name} - Actual vs Predicted:\n")
    print(actual_vs_predicted[name].head()) 

    print('------------------------------------------------------')
    print(f"\n{name} - Classification Report:\n")
    print(classification_report(y_test, y_pred))

# Convert validation results to DataFrame
metrics_df = pd.DataFrame(validation_results).T
print("\nValidation Metrics Summary:\n", metrics_df)
print(' ')


# Save actual vs predicted values for each model
for name, df in actual_vs_predicted.items():
    df.to_csv(f"./data/{name}_actual_vs_predicted_diabetes.csv", index=False)
    print(f'Validation results of {name} have been saved to ./data/{name}_actual_vs_predicted_diabetes.csv')
    
print(' ')

````
