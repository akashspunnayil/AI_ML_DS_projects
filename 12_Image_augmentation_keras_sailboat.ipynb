```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import img_to_array, load_img, smart_resize

# Load an image from file
image_path = './data/ship.jpeg'  # Replace with the actual image path
image = load_img(image_path)  # Load image (any size)
image = img_to_array(image)  # Convert to array

# Resize the image using smart_resize
size = (300, 300)  # Define the target size (width, height)
resized_image = smart_resize(image, size)

# Display the original and resized images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image.astype('uint8'))
plt.title("Original Image")
plt.axis('off')

# Resized image
plt.subplot(1, 2, 2)
plt.imshow(resized_image.astype('uint8'))
plt.title(f"Resized Image to {size}")
plt.axis('off')

plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.preprocessing.image import apply_affine_transform

# Load an image from file
# image_path = 'cat/img.jpg'  # Replace with your image path
image = load_img(image_path)  # Load image (any size)
image = img_to_array(image)  # Convert to array

# Define parameters for the affine transformation
theta = 20  # Rotation angle in degrees
tx = 10     # Translation along x-axis
ty = 5      # Translation along y-axis
shear = 0   # Shear intensity
zx = 1.2    # Zoom factor along x-axis
zy = 1.2    # Zoom factor along y-axis

# Apply affine transformation
transformed_image = apply_affine_transform(
    image,
    theta=theta,
    tx=tx,
    ty=ty,
    shear=shear,
    zx=zx,
    zy=zy,
    fill_mode='nearest'
)

# Display the original and transformed images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image.astype('uint8'))
plt.title("Original Image")
plt.axis('off')

# Transformed image
plt.subplot(1, 2, 2)
plt.imshow(transformed_image.astype('uint8'))
plt.title("Transformed Image")
plt.axis('off')

plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, NumpyArrayIterator
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Load a single image from file (replace 'your_image.jpg' with your actual image path)
# image_path = 'cat/img.jpg'  # Adjust the path to your image
image = load_img(image_path, target_size=(800, 800))  # Load the image and resize
image_array = img_to_array(image)  # Convert to a NumPy array

# Add a new axis to make it a batch of one image
image_array = np.expand_dims(image_array, axis=0)

# Create a dummy label for the image
dummy_label = np.array([0])  # You can set this to any value you want

# Create an ImageDataGenerator instance for data augmentation
image_data_generator = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Create a NumpyArrayIterator for the single image
iterator = NumpyArrayIterator(
    x=image_array,
    y=dummy_label,
    image_data_generator=image_data_generator,
    batch_size=1,  # Batch size of 1 for a single image
    shuffle=True,  # Shuffle the data
    seed=42,
    data_format='channels_last'
)

# Generate augmented images
augmented_images = []
for _ in range(4):  # Generate 4 augmented images
    augmented_images.append(next(iterator)[0])  # Use next(iterator) to get augmented images

# Display the original and augmented images
plt.figure(figsize=(10, 10))

# Original image
plt.subplot(3, 2, 1)
plt.imshow(image_array[0].astype('uint8'))
plt.title("Original Image")
plt.axis('off')

# Augmented images
for i, augmented_image in enumerate(augmented_images):
    plt.subplot(3, 2, i + 2)
    plt.imshow(augmented_image[0].astype('uint8'))  # Get the single image from the batch
    plt.title(f"Augmented Image {i + 1}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
